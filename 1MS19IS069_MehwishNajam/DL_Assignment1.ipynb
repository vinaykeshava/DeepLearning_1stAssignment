ADDING THE COLUMN NAMES

```python
df.columns = ['C' + str(n) for n in range(1,29)]
```

REMOVING THE NON-NUMERICAL VALUES

```python
df.drop(['C1'] + ['C' + str(n) for n in range(20,29)], axis=1, inplace = True)
df.head()
```

```python
df.median(axis = 0)
```

```python
plt.rcParams['figure.figsize'] = [50, 50]
```

NORMALISATION / MINMAXSCALING

```python
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
NDf = min_max_scaler.fit_transform(df)

NDf = pd.DataFrame(NDf)

for i in range(18):
  m = np.mean(NDf[i])
  cols = np.where((NDf[i] >= m - (0.8 * m)) & (NDf[i] <= m + (0.8 * m)), 'm', 'b')
  plt.subplot(4,5,i + 1)
  plt.scatter(NDf.index,NDf[i], c = cols)
plt.show()
```

PLOTTING STANDARD SCALER / Z - SCORE

```python
from sklearn import preprocessing
z_scaler = preprocessing.StandardScaler()
NDf = z_scaler.fit_transform(df)

NDf = pd.DataFrame(NDf)

for i in range(18):
  m = np.mean(NDf[i])
  cols = np.where((NDf[i] >= -0.8) & (NDf[i] <= 0.8), 'm', 'b')
  plt.subplot(4,5,i + 1)
  plt.scatter(NDf.index,NDf[i], c = cols)
plt.show()
```

PLOTTING MAX ABSOLUTE VALUE

```python
from sklearn import preprocessing
z_scaler = preprocessing.MaxAbsScaler()
NDf = z_scaler.fit_transform(df)

NDf = pd.DataFrame(NDf)

for i in range(18):
  m = np.mean(NDf[i])
  cols = np.where((NDf[i] >= m - (0.8 * m)) & (NDf[i] <= m + (0.8 * m)), 'm', 'b')
  plt.subplot(4,5,i + 1)
  plt.scatter(NDf.index,NDf[i], c = cols)
plt.show()
```
