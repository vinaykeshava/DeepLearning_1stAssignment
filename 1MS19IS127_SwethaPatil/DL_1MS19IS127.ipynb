```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
df = pd.read_fwf("/17.txt", header = None)
```

```
df.head()
```

Adding Column Name

```
df.columns = ['c' + str(n) for n in range(1,29)]
```

```
df.head()
```

Removing non numerical values

```
df.drop(['c1'] + ['c' + str(n) for n in range(20,29)], axis=1, inplace = True)
df.head()
```

Finding Mean and other values

```
df.describe()
```

Finding Median

```
df.median(axis = 0)
```

```
plt.rcParams['figure.figsize'] = [50, 50]
```

Plotting MinMaxScaler

```
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
newDf = min_max_scaler.fit_transform(df)

newDf = pd.DataFrame(newDf)

for i in range(18):
  m = np.mean(newDf[i])
  cols = np.where((newDf[i] >= m - (0.8 * m)) & (newDf[i] <= m + (0.8 * m)), 'b', 'r')
  plt.subplot(4,5,i + 1)
  plt.scatter(newDf.index,newDf[i], c = cols)
plt.show()
```

Plotting Standard Scaler / Z score

```
from sklearn import preprocessing
z_scaler = preprocessing.StandardScaler()
newDf = z_scaler.fit_transform(df)

newDf = pd.DataFrame(newDf)

for i in range(18):
  m = np.mean(newDf[i])
  cols = np.where((newDf[i] >= -0.8) & (newDf[i] <= 0.8), 'b', 'r')
  plt.subplot(4,5,i + 1)
  plt.scatter(newDf.index,newDf[i], c = cols)
plt.show()
```

Plotting Max Absolute

```
from sklearn import preprocessing
z_scaler = preprocessing.MaxAbsScaler()
newDf = z_scaler.fit_transform(df)

newDf = pd.DataFrame(newDf)

for i in range(18):
  m = np.mean(newDf[i])
  cols = np.where((newDf[i] >= m - (0.8 * m)) & (newDf[i] <= m + (0.8 * m)), 'b', 'r')
  plt.subplot(4,5,i + 1)
  plt.scatter(newDf.index,newDf[i], c = cols)
plt.show()
```
