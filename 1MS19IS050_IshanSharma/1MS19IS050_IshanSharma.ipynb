# Maximum Absolute Scaling

```python
from sklearn import preprocessing
z_scaler = preprocessing.MaxAbsScaler()
newDf = z_scaler.fit_transform(df)

newDf = pd.DataFrame(newDf)

for i in range(18):
  m = np.mean(newDf[i])
  cols = np.where((newDf[i] >= m - (0.8 * m)) & (newDf[i] <= m + (0.8 * m)), 'm', 'k')
  plt.subplot(4,5,i + 1)
  plt.scatter(newDf.index,newDf[i], c = cols)
plt.show()
```

# MinMax Scaling

```python
min_max_scaler = preprocessing.MinMaxScaler()
df_minmax = min_max_scaler.fit_transform(df)

df_minmax = pd.DataFrame(df_minmax)

for i in range(19):
  m = np.mean(df_minmax[i])
  columns = np.where((df_minmax[i] >= m - (0.8 * m)) & (df_minmax[i] <= m + (0.8 * m)), 'm', 'k')
  plt.subplot(4,5,i + 1)
  plt.scatter(df_minmax.index,df_minmax[i], c = columns)
plt.show()
```

# Z-Score OR Standard Score

```python
z_scaler = preprocessing.StandardScaler()
df_z = z_scaler.fit_transform(df)

df_z = pd.DataFrame(df_z)

for i in range(18):
  m = np.mean(df_z[i])
  cols = np.where((df_z[i] >= -0.8) & (df_z[i] <= 0.8), 'm', 'k')
  plt.subplot(4,5,i + 1)
  plt.scatter(df_z.index,newDf[i], c = cols)
plt.show()
```
