Plotting MinMaxScaler

```python
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
newDf = min_max_scaler.fit_transform(data)

newDf = pd.DataFrame(newDf)

for i in range(18):
  m = np.mean(newDf[i])
  cols = np.where((newDf[i] >= m - (0.8 * m)) & (newDf[i] <= m + (0.8 * m)), 'g', 'r')
  plt.subplot(4,5,i + 1)
  plt.scatter(newDf.index,newDf[i], c = cols)
plt.show()
```

Plotting Standard Scaler / Z score

```python
from sklearn import preprocessing
z_scaler = preprocessing.StandardScaler()
newDf = z_scaler.fit_transform(data)

newDf = pd.DataFrame(newDf)

for i in range(18):
  m = np.mean(newDf[i])
  cols = np.where((newDf[i] >= -0.8) & (newDf[i] <= 0.8), 'g', 'r')
  plt.subplot(4,5,i + 1)
  plt.scatter(newDf.index,newDf[i], c = cols)
plt.show()
```
